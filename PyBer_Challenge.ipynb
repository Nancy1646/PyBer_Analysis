### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
#city_data_to_load = pd.read_csv("Resources/city_data.csv")
#ride_data_to_load = pd.read_csv("Resources/ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv("Resources/city_data.csv")
ride_data_df = pd.read_csv("Resources/ride_data.csv")
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
#pyber_data_df.groupby("type")
pyber_data_df_ride_id = pyber_data_df.groupby("type").count()
pyber_data_df_ride_id = pyber_data_df_ride_id.loc[:, "ride_id"]
pyber_data_df_ride_id
```

```python
# 2. Get the total drivers for each city type
pyber_data_df_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
#pyber_data_df_driver_count = pyber_data_df_driver_count.loc[:, "driver_count"]
pyber_data_df_driver_count
```

```python
#  3. Get the total amount of fares for each city type
pyber_data_df_fare = pyber_data_df.groupby("type").sum("fare")
pyber_data_df_fare = pyber_data_df_fare.loc[:, "fare"]
pyber_data_df_fare
```

```python
#  4. Get the average fare per ride for each city type. 
avg_per_ride = pyber_data_df_fare/pyber_data_df_ride_id
avg_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_per_driver = pyber_data_df_fare/pyber_data_df_driver_count
avg_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.concat([pyber_data_df_ride_id, pyber_data_df_driver_count, pyber_data_df_fare, avg_per_ride, avg_per_driver], axis = 1)
pyber_summary_df = pyber_summary_df.rename({'ride_id':'Total Rides', 'driver_count':'Total Drivers', 'fare': 'Total Fares', 0:'Average Fare Per Ride', 1:'Average Fare Per Driver'}, axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.


pyber_summary_df.loc[:,"Total Rides"] = pyber_summary_df["Total Rides"].map('{:,}'.format)
pyber_summary_df.loc[:,"Total Drivers"] = pyber_summary_df["Total Drivers"].map('{:,}'.format)
pyber_summary_df.loc[:,"Total Fares"] = pyber_summary_df["Total Fares"].map('${:,.2f}'.format)
pyber_summary_df.loc[:,"Average Fare Per Ride"] = pyber_summary_df["Average Fare Per Ride"].map('${:,.2f}'.format)
pyber_summary_df.loc[:,"Average Fare Per Driver"] = pyber_summary_df["Average Fare Per Driver"].map('${:,.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares = pyber_data_df.groupby(["date", "type"]).sum()["fare"]
sum_fares
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_fares = sum_fares.reset_index()
sum_fares
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
#pd.pivot_table(df, values = 'Value', index=['Country','Year'], columns = 'Indicator').reset_index()

pivot_table = sum_fares.pivot(index = 'date', columns = 'type', values = 'fare')
pivot_table.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

Jan_Apr = pivot_table.loc['2019-01-01':'2019-04-29']
Jan_Apr
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
Jan_Apr.index = pd.to_datetime(Jan_Apr.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
Jan_Apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_df = Jan_Apr.resample('W').sum()
weekly_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 


ax = weekly_df.plot(figsize = (20,10))

ax.set_ylabel("Fare ($USD)")
ax.set_xlabel("")
ax.set_title("Total Fare By City Type")
plt.legend(bbox_to_anchor=(1.0, 1.0))

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.savefig('analysis/PyBer_fare_summary.png', bbox_inches = 'tight')
```
